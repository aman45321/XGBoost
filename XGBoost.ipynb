```python
X=df.iloc[:,:-1]
X
```

```python
y=df['Is Diabetic']
y
```

```python
scale=StandardScaler()
X_scale=scale.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_scale,y,test_size=0.30,random_state=42)
```

```python
model=XGBClassifier(objective='binary:logistic')
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_train)
prediction=[round(value) for value in y_predict]
accuracy=accuracy_score(y_train,prediction)
accuracy
```

```python
x_predict=model.predict(x_test)

prediction=[round(value) for value in x_predict]
accuracy=accuracy_score(y_test,prediction)
accuracy
```

```python
param_grid={
   
    ' learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[10,50,100,200]
    
}
```

```python
grid_search=GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid,verbose=3)
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
new_model=XGBClassifier(learnng_rate=1,max_depth=6,n_estimators=50)
new_model.fit(x_train,y_train)
```

```python
y_predict=new_model.predict(x_test)
prediction=[round(value) for value in y_predict]
accuracy=accuracy_score(y_test,prediction)
accuracy
```
